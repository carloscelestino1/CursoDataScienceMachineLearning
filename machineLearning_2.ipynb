```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_logica = pd.DataFrame(np.random.randint(28, size=(100, 1)), columns=['log_acertos'])
df_logica
```

```python
df_portugues = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['port_acertos'])
df_portugues
```

```python
df_java = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['java_acertos'])
df_java
```

```python
df_python = pd.DataFrame(np.random.randint(40, size=(100,1)), columns=['py_acertos'])
df_python
```

```python
df_formacao = pd.DataFrame(np.random.randint(2, size=(100, 1)), columns=['formacao'])
df_formacao
```

```python
df = pd.concat([df_logica,df_portugues,df_java,df_python,df_formacao], axis=1)
```

```python
df.head(30)
```

```python
x = df[['log_acertos', 'port_acertos', 'java_acertos', 'py_acertos']]
```

```python
y = df[['formacao']]
```

```python
from sklearn.model_selection import train_test_split

```

```python
train_test_split(x,y, test_size=0.25)
```

```python
treino_x, teste_x, treino_y, teste_y = train_test_split(x,y, test_size=0.25)
print(treino_x.shape)
print(teste_x.shape)
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

```python
treino_x, teste_x, treino_y, teste_y = train_test_split(x,y, test_size=0.25)
print("Treinaremos com %d elementos e testaremos com %d elementos" %(len(treino_x), len(teste_x)))
```

```python
modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
```

```python
print('A acuracia foi %.2f%%' % acuracia)
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

SEED = 20

treino_x, teste_x, treino_y, teste_y = train_test_split(x,y, random_state=SEED, test_size=0.25)
print("Treinaremos com %d elementos e testaremos com %d elementos" %(len(treino_x), len(teste_x)))

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100

print('A acuracia foi %.2f%%' % acuracia)
```
