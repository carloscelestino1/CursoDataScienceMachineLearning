```python
import matplotlib.pyplot as plt

x = [1,2,5]
y = [2,3,7]

#titulo
plt.title("gráfico em linha")

#eixos
plt.xlabel("eixo x")
plt.ylabel("eixo y")

plt.plot(x,y)
plt.show
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,3,7,1,0]

#titulo
plt.title("gráfico em barras")

#eixos
plt.xlabel("eixo x")
plt.ylabel("eixo y")

plt.bar(x,y)
plt.show
```

```python
import matplotlib.pyplot as plt

x1 = [1,3,5,7,9]
y1 = [2,3,7,1,0]

x2 = [2,4,6,8,10]
y2 = [5,1,3,7,4]


plt.title("gráfico em barras 2")
plt.xlabel("eixo x")
plt.ylabel("eixo y")

plt.bar(x1,y1, label = 'Grupo 1') #adicionando legendas
plt.bar(x2,y2, label = 'Grupo 2')
plt.legend()

plt.show
```
