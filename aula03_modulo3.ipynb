```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
empresas = ['Apple', 'Google', 'Amazon', 'Microsoft', 'Tecent', 'McDonalds', 'Visa', 'Facebook', 'Alibaba', 'Louis Vuitton']
```

```python
bilhoes = [947.062, 819.573, 705.646, 611.460, 214.023, 196.526, 191.032, 186.421, 169.966, 124.273]
```

```python
plt.plot(empresas, bilhoes)
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
```

```python
plt.xlabel('Empresas')
plt.ylabel('Bilh√µes')
plt.title('Empresas mais ricas do mundo')
plt.plot(empresas, bilhoes)
```

```python
plt.rcParams['figure.figsize'] = (15,8)
plt.xlabel('Empresas')
plt.ylabel('Bilh√µes')
plt.title('Empresas mais ricas do mundo')
plt.plot(empresas, bilhoes)
plt.show()
```

```python

plt.xlabel('Empresas')
plt.ylabel('Bilh√µes')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.plot(empresas, bilhoes)
plt.show()
```

```python
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilh√µes', fontsize=20)
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.plot(empresas, bilhoes, color = 'navy',marker='o')
plt.legend(['Bilh√µes de dolares'], loc='right', fontsize=15)
plt.grid(True)
plt.show()
```

```python
plt.bar(empresas, bilhoes, color='green')
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilh√µes', fontsize=20)
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.legend(['Bilh√µes de dolares'], loc='high right', fontsize=15)
plt.grid(True)
plt.show()
```
