```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
empresas = ['Apple', 'Google', 'Amazon',
            'Microsoft', 'Tencent',
            'McDonalds', 'Visa',
            'Facebook', 'Alibaba',
            'Louis vuiton']
```

```python
bilhoes = [947.062, 819.573, 705.646,
           611.460, 214.023, 196.526,
           191.032, 186.421, 169.966,
           124.273]
```

```python
plt.plot(empresas,bilhoes)
plt.title("Empresas mais ricas do mundo")
```

```python
plt.plot(empresas, bilhoes)
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
plt.title("Empresas mais ricas do mundo")
```

```python
plt.rcParams['figure.figsize'] = [15,8]
plt.plot(empresas, bilhoes)
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
plt.title("Empresas mais ricas do mundo")
plt.show()
```

```python
plt.plot(empresas, bilhoes)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title("Empresas mais ricas do mundo", fontsize=25)
plt.show()
```

```python
plt.plot(empresas, bilhoes, color='green')
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title("Empresas mais ricas do mundo", fontsize=25)
plt.legend(['Bilhões de Dólares'], loc='right', fontsize=15)
plt.show()
```

```python
plt.plot(empresas, bilhoes, color='green', lw=10) #aumentando expessura da linha
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title("Empresas mais ricas do mundo", fontsize=25)
plt.legend(['Bilhões de Dólares'], loc='right', fontsize=15)
plt.show()
```

```python
plt.plot(empresas, bilhoes, color='green', lw=2, marker='o') #aplicando marcardor
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title("Empresas mais ricas do mundo", fontsize=25)
plt.legend(['Bilhões de Dólares'], loc='right', fontsize=15)
plt.show()
```

```python
plt.plot(empresas, bilhoes, color='green', lw=2, marker='o')
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title("Empresas mais ricas do mundo", fontsize=25)
plt.legend(['Bilhões de Dólares'], loc='right', fontsize=15)
plt.grid(True) #adicionando linha de grade
plt.show()
```

```python
plt.bar(empresas, bilhoes, color='green')
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title("Empresas mais ricas do mundo", fontsize=25)
plt.legend(['Bilhões de Dólares'], loc='high right', fontsize=15) #mudança de local da legenda
plt.grid(True)
plt.show()
```

```python
plt.bar(empresas, bilhoes)
```

```python
cores = ['red', 'yellow', 'green', 'blue', 'purple', 'orange', 'magenta', 'white', 'brown', 'black']
```

```python
plt.bar(empresas, bilhoes, color= cores, edgecolor='black') #adicionando cores
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title("Empresas mais ricas do mundo", fontsize=25)
plt.legend(['Bilhões de Dólares'], loc='high right', fontsize=15) #mudança de local da legenda
plt.grid(True)
plt.show()
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-dark') #modificando estilo
plt.bar(empresas, bilhoes, color= cores, edgecolor='black')
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title("Empresas mais ricas do mundo", fontsize=25)
plt.legend(['Bilhões de Dólares'], loc='high right', fontsize=15) 
plt.grid(True)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [15,8]
plt.style.use('seaborn-dark') 
plt.barh(empresas, bilhoes, color= cores, edgecolor='black') #grafico invertido
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title("Empresas mais ricas do mundo", fontsize=25)
plt.legend(['Bilhões de Dólares'], loc='high right', fontsize=15) 
plt.grid(True)
plt.show()
```

```python
plt.style.use('seaborn-dark') 
plt.scatter(empresas, bilhoes, color= cores, edgecolor='black')

plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title("Empresas mais ricas do mundo", fontsize=25)
plt.legend(['Bilhões de Dólares'], loc='high right', fontsize=15) 
plt.grid(True)
plt.show()
```
