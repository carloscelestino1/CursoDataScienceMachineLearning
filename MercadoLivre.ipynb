# Buscando produtos no Mercado Livre

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url_base = 'https://lista.mercadolivre.com.br/'
```

```python
produto_nome = input('Qual produto vocÃª procura? ')
```

```python
response = requests.get(url_base + produto_nome)
print(response)
```

```python
response.text
```

```python
site = BeautifulSoup(response.text, 'html.parser') #faz a requisiÃ§Ã£o passando  o  formato html
```

```python
site
```

```python
#procurar pela tag div e os atributos attrs

produto = site.find('div', attrs= {'class' : 'ui-search-result__wrapper'})
```

```python
print(produto)
```

```python
links = produto.find('a', attrs={'class', 'ui-search-link'})
```

```python
print('Link do produto: ', links['href'])
```

```python
#localizando o preÃ§o

real= produto.find('', attrs={'class': 'price-tag-fraction'})
```

```python
print('PreÃ§o do produto: R$ ', real.text)
```

# Buscando todos os produtos

```python
#criando um script

from numpy.lib.function_base import append
produto_nome=input('qual produto vocÃª procura? ')
produtos= site.findAll('div', attrs= {'class':'ui-search-result__wrapper'})


a = []
for produto in produtos:
  
  links = produto.find('a', attrs={'class', 'ui-search-link'})
  real= produto.find('', attrs={'class': 'price-tag-fraction'})
  print('Link do produto: ', links['href'])
  print('PreÃ§o do produto: R$ ', real.text)
  valor= real.text
  a+=[valor]
  
  print('\n')
```

```python
dados = pd.DataFrame(a)
dados.columns = ['PreÃ§o']
```
